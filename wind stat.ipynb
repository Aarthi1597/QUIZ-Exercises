```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("wind.data",sep=" ")
```

```python
time = data[['Yr','Mo','Dy']]
time.columns = ['Year','Month','Day']
ne = pd.DataFrame([])
```

```python
date_t = ne.append(time)

def Year(x):
    return (1900 + x)

date_t['Year']=date_t['Year'].apply(Year)
date_t['Month'] = date_t['Month'].fillna(1)
date_t['Year'] = date_t['Year'].fillna(0)
date_t['Day'] = date_t['Day'].fillna(1)
date_t['Year'] = date_t['Year'].apply(int)
date_t['Month'] = date_t['Month'].apply(int)
date_t['Day'] = date_t['Day'].apply(int)
```

```python
date_t['Date_time'] = pd.DataFrame(date_t['Day'].map(str) + '-'+ date_t['Month'].map(str) + '-' + date_t['Year'].map(str))
date_t['Date_time'] = pd.to_datetime(date_t['Date_time'])
```

```python
data = data.join(date_t)
```

```python
data = data.drop("Mo",axis=1)
data = data.drop("Dy",axis=1)
data = data.drop("Yr",axis=1)
data = data.drop("Year",axis=1)
data = data.drop("Month",axis=1)
data = data.drop("Day",axis=1)
```

## Downsample the record to monthly record for each location

```python
data['trans_date'] = data['Date_time'].dt.strftime('%B - %Y')
```

```python
data
```

```python
m_data = data[['RPT','VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL','trans_date']].fillna(0)
```

```python
data_pivot =  m_data.pivot_table(index = 'trans_date',aggfunc=sum,fill_value=0).T
```

```python
data_pivot
```

## Calculate mean wind speed for each month in dataset

```python
m = pd.DataFrame(data_pivot.mean())
m.columns = ['Month_Mean']
```
