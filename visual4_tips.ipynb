### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x = tips['time'], y = tips['total_bill'])
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
def Din(x):
    if(x == 'Dinner'):
        return "1"
    else:
        return np.nan
tips['Dinner'] = tips['time'].apply(Din)


def Lun(x):
    if(x == 'Lunch'):
        return "1"
    else:
        return np.nan
tips['Lunch'] = tips['time'].apply(Lun)
```

```python
sns.distplot(a = tips['Dinner'],kde = False,bins = 100)
```

```python
plt.figure(figsize = (16,6))
sns.distplot(a = tips['tip'],kde = False,bins = 100)
#kde = False,bins = 100
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
def sexF(x):
    if(x == 'Female'):
        return "Female"
    else:
        return 0
tips['Female'] = tips['sex'].apply(sexF)
```

```python

def sexM(x):
    if(x == 'Male'):
        return "Male"
    else:
        return 0
tips['Male'] = tips['sex'].apply(sexM)
```

```python
plt.figure(figsize = (6,6))
sns.scatterplot(x = tips['total_bill'],y = tips['tip'],hue = tips['smoker'],style = tips['Male'])

plt.figure(figsize = (6,6))
sns.scatterplot(x = tips['total_bill'],y = tips['tip'],hue = tips['smoker'],style = tips['Female'])
```
